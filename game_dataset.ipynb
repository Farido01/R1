```python

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
%config InlineBackend.figure_fornmat = 'svg'

from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
import  pandas as pd
```

```python
df = pd.read_csv('game_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df['User_Score'] = df.User_Score.astype('float64')
df['Year_of_Release'] = df.Year_of_Release.astype('int64')
df['User_Count'] = df.User_Count.astype('int64')
```

```python
df.head()
```

```python
df['Critic_Score'] = df.Critic_Score.astype('int64')
```

```python
df.shape
```

```python
df.head()
```

```python
useful_cols = ['Name', 'Platform', 'Year_of_Release', 'Genre', 'Global_Sales', 'Critic_Score', 'Critic_Count',
               'User_Score', 'User_Count', 'Rating']
df[useful_cols].head()
```

```python
sales_df=df[[x for x in df.columns if 'Sales' in x]+['Year_of_Release']]
sales_df.groupby('Year_of_Release').sum().plot()
```

```python
sales_df.groupby('Year_of_Release').sum().plot(kind='bar', rot=45)
```

```python
cols = ['Global_Sales', 'Critic_Score', 'Critic_Count',
               'User_Score', 'User_Count']
sns_plot = sns.pairplot(df[cols])
sns_plot.savefig('pairplot.png')
```

```python
sns.distplot(df['Critic_Score'])
```

```python
top_platforms = df['Platform'].value_counts().sort_values(ascending = False).head(5).index.values
```

```python
import seaborn as sns
sns.boxplot(y='Platform', x='Critic_Score', data=df[df['Platform'].isin(top_platforms)], orient='h', hue='Platform')
```

```python
import seaborn as sns

sns.boxplot(y='Platform', x='Critic_Score', data=df[df['Platform'].isin(top_platforms)], orient='h', palette='Set2')
```

```python
platform_genre_sales=df.pivot_table(
                        index = 'Platform',
                        columns='Genre',
                        values='Global_Sales',
                        aggfunc=sum).fillna(0).applymap(float)
sns.heatmap(platform_genre_sales,annot=True,fmt='.1f', linewidths=.5)
```
